## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

Create a scatter plot of `Age` and `WorkWeekHrs.`

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUEARY = '''
SELECT * FROM DatabaseWorkedWith
'''
dbDF = pd.read_sql_query(QUEARY, con=conn)

# dbDF.DatabaseWorkedWith.value_counts().to_frame()
ms = dbDF[dbDF.DatabaseWorkedWith == 'MySQL']

notOnlySql = dbDF[(dbDF.Respondent.isin(ms.Respondent))]

filterRes = notOnlySql[(notOnlySql.DatabaseWorkedWith != 'MySQL')]
ResNum = filterRes.Respondent
newMs = ms[~ms.Respondent.isin(ResNum.values)]
len(newMs)
```

```python
import matplotlib.pyplot as plt
QUEARY = '''
SELECT * FROM DatabaseDesireNextYear
'''
dbDF = pd.read_sql_query(QUEARY, con=conn)

firstFive = dbDF.DatabaseDesireNextYear.value_counts().to_frame()[:5]
allDB = dbDF.DatabaseDesireNextYear.value_counts().sum()

firstFive.DatabaseDesireNextYear.plot.pie(
    figsize=(12, 4),
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    labels=None,
    pctdistance=1.12,
)
plt.axis('equal') 

plt.legend(labels = firstFive.index, loc='upper left')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
tempDf = df[['WorkWeekHrs', 'CodeRevHrs', 'Age']].set_index('Age')
tempDf = tempDf[(tempDf.index > 30) & (tempDf.index < 35)]
tempDf.index.value_counts()
tempGr = tempDf.groupby(['Age'])['WorkWeekHrs', 'CodeRevHrs'].median()

tempGr.plot.bar(stacked = True)
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
tempDf = df[['ConvertedComp', 'Age']].set_index('Age')
tempDf = tempDf[(tempDf.index > 45) & (tempDf.index < 60)]
tempGr = tempDf.groupby(['Age'])['ConvertedComp'].median()

tempGr.plot.line()
```
